#Missing Values Handling

Features with extremely high missing ratio (>90%) were filled with zero and
missing indicators were created.

Credit history related variables were imputed with median values and
missing flags.

Credit activity variables were filled with zero assuming absence of activity.

Categorical variables were filled with the most frequent category.

Low missing numerical variables were filled with median values.

```python
very_sparse = [
"policyrule_partialdecision",
"swo_ind_overdue_loans_number",
"swo_ind_same_ip_different_pesel_count",
"swo_ind_same_ip_different_pesel_count_unique"
]

for col in very_sparse:

    train[col+"_missing"] = train[col].isna().astype(int)
    testdf[col+"_missing"] = testdf[col].isna().astype(int)

    train[col] = train[col].fillna(0)
    testdf[col] = testdf[col].fillna(0)
```

```python
credit_history_cols = [
"days_after_knownsince_pbd",
"max_dpd",
"avg_dpd",
"daysafterregistration"
]

for col in credit_history_cols:

    train[col+"_missing"] = train[col].isna().astype(int)
    testdf[col+"_missing"] = testdf[col].isna().astype(int)

    median = train[col].median()

    train[col] = train[col].fillna(median)
    testdf[col] = testdf[col].fillna(median)
```

```python
from sklearn.metrics import roc_auc_score
import numpy as np
import pandas as pd

# using only numerical features
num_cols = train.select_dtypes(include=["int64","float64"]).columns
num_cols = num_cols.drop("TARGET")

auc_scores = []

for col in num_cols:

    # remove NaN for calculation
    tmp = train[[col, "TARGET"]].dropna()

    if tmp[col].nunique() > 1:
        auc = roc_auc_score(tmp["TARGET"], tmp[col])
        auc_scores.append((col, auc))

auc_df = pd.DataFrame(auc_scores, columns=["feature","auc"])

# features where auc is far from 0.5
auc_df["auc_abs"] = abs(auc_df["auc"] - 0.5)

top_features = auc_df.sort_values("auc_abs", ascending=False).head(10)

top_features
```

Variables describing positive credit history (e.g. number of prepaid loans,

Several variables exhibited strong right skewness and significant outliers.
Logarithmic transformation (log1p) was applied to stabilize the variance

```python
def ks_statistic(y_true, y_pred):

    df = pd.DataFrame({
        "y": y_true,
        "score": y_pred
    })

    df = df.sort_values("score")

    df["good"] = 1 - df["y"]
    df["bad"] = df["y"]

    df["cum_good"] = df["good"].cumsum() / df["good"].sum()
    df["cum_bad"] = df["bad"].cumsum() / df["bad"].sum()

    ks = np.max(np.abs(df["cum_bad"] - df["cum_good"]))

    return ks
```

```python
ks = ks_statistic(y_val, pred_lgb)

print("KS statistic:", ks)
```

```python
def plot_ks(y_true, y_pred):

    df = pd.DataFrame({
        "y": y_true,
        "score": y_pred
    }).sort_values("score")

    df["good"] = 1 - df["y"]
    df["bad"] = df["y"]

    df["cum_good"] = df["good"].cumsum() / df["good"].sum()
    df["cum_bad"] = df["bad"].cumsum() / df["bad"].sum()

    df["population"] = np.arange(1, len(df)+1) / len(df)

    ks = np.max(np.abs(df["cum_bad"] - df["cum_good"]))

    plt.figure(figsize=(6,4))
    plt.plot(df["population"], df["cum_good"], label="Good")
    plt.plot(df["population"], df["cum_bad"], label="Bad")

    plt.title(f"KS Curve (KS = {ks:.3f})")
    plt.xlabel("Population")
    plt.ylabel("Cumulative Distribution")

    plt.legend()
    plt.show()

    return ks
```

```python
ks = plot_ks(y_val, pred_lgb)

print("KS statistic:", ks)
```

KS Statistic

The KolmogorovâSmirnov (KS) statistic was used to evaluate the modelâs
ability to separate defaulted and non-defaulted clients.

The KS value obtained for the validation dataset is 0.483, indicating
strong discriminatory power. The model is able to clearly distinguish
between higher-risk and lower-risk borrowers.



```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score
import numpy as np

kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

auc_scores = []

for train_idx, val_idx in kf.split(X, y):

    X_tr, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_tr, y_val = y.iloc[train_idx], y.iloc[val_idx]

    model = lgb.LGBMClassifier(
        n_estimators=300,
        learning_rate=0.05,
        max_depth=5,
        random_state=42
    )

    model.fit(X_tr, y_tr)

    pred = model.predict_proba(X_val)[:,1]

    auc = roc_auc_score(y_val, pred)

    auc_scores.append(auc)

print("Mean AUC:", np.mean(auc_scores))
print("Std AUC:", np.std(auc_scores))
```

To evaluate the robustness of the model, 5-fold stratified cross-validation
was performed.

The model achieved a mean ROC-AUC score of approximately 0.81 with a
standard deviation of 0.008.

These results indicate that the model demonstrates strong predictive
performance and stable behaviour across different folds of the data,
suggesting good generalization ability.
